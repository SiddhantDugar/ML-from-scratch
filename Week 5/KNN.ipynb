## K_Nearest_Neighbours

## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Importing the dataset

```python
Xtrain = pd.read_csv('Diabetes_XTrain.csv')
ytrain = pd.read_csv('Diabetes_YTrain.csv')
Xtest  = pd.read_csv('Diabetes_Xtest.csv')
X_train = Xtrain.iloc[:, :].values
y_train = ytrain.iloc[:, :].values
X_test = Xtest.iloc[:, :].values
```

## Dataset visualisation

```python
plt.figure(figsize=(9,5))
sns.heatmap(Xtrain.corr(),annot=True,linewidth = 0.5, cmap='coolwarm')
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
print(X_test)
```

## Training the K-NN model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(y_pred.reshape(len(y_pred),1))
```
